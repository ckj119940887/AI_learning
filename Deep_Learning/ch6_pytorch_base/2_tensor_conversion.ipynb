```python
import numpy as np
import torch
```

```python
# 张量元素类型转换
tensor0 = torch.tensor([1,2,3])
print(tensor0.dtype)
tensor1 = tensor0.type(torch.float64)
print(tensor1.dtype)
```

```python
# 直接转换
tensor2 = tensor0.int() # bool(), float(), double()
```

```python
# tensor --> ndarray
tensor2 = torch.rand(3, 2)
np_array1 = tensor2.numpy()
print(np_array1)

# tensor与numpy共享内存
tensor2[:, 0] = 10
print(tensor2)
print(np_array1)

# 使用copy方法避免共享内存
np_array2 = tensor2.numpy().copy()
tensor2[:, 0] = 100
print(tensor2)
print(np_array2)
```

```python
# ndarray --> tensor,仍是共享内存
np_array = np.random.randn(3,2)
tensor3 = torch.from_numpy(np_array)
print(np_array)
print(tensor3)
```

```python
# 使用tensor方法，将ndarray转换为tensor,不共享内存了
np_array = np.random.randn(3)
tensor4 = torch.tensor(np_array)
np_array[0] = 100
print(np_array)
print(tensor4)
```
