```python
import torch
```

```python
tensor1 = torch.randint(low=0, high=10, size=(3,2,4))

print(tensor1)
print(tensor1[0, 1, 2])
```

```python
# 所有矩阵第二行
print(tensor1[:, 1])
```

```python
# 范围索引
print(tensor1[1:])
print(tensor1[1:, 0:, 2:3])

# 支持步长
print(tensor1[1:, 0:, 2:3:1])
```

```python
# 列表索引
# 对应匹配，相当于0矩阵第0行，1矩阵第1行
print(tensor1[[0, 1], [0, 1]])

# 0矩阵的前两行，1矩阵的前两行
print(tensor1[[[0], [1]], [0, 1]])
```

```python
# 布尔索引
mask = tensor1 > 5
print(mask)
print(tensor1[mask])

# 选取首元素大于5的所有行
mask2 = tensor1[:, :, 0] > 5
print(mask2)
print(tensor1[mask2])
```

```python
# 选取第二个元素大于5的所有列
mask = tensor1[:, 1, :] > 5
print(mask)
#会报错,shape mismatch
#print(tensor1[mask])

#转换维度
# 0维不变，1维和2维互换
tensor2 = tensor1.permute(0, 2, 1)
print(tensor2)
print(tensor2[mask].T)
```
