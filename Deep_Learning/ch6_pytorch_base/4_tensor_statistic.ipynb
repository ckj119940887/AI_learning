```python
import torch
```

```python
#常见运算函数：
#sum()求和
#mean()求均值
#max()/min()求最大/最小值及其索引
#argmax()/argmin()求最大值/最小值的索引
#std()求标准差
#unique()去重
#sort()排序
```

```python
tensor1 = torch.randint(low=0, high=10, size=(3, 2, 4))
print(tensor1)

print("sum:", tensor1.sum())
print("sum:", tensor1.sum(dim=0))

"""
以如下高维数据为例：
tensor([[[3, 0, 0, 5],
         [2, 7, 6, 1]],

        [[8, 2, 5, 6],
         [3, 4, 4, 8]],

        [[7, 5, 5, 9],
         [2, 7, 3, 7]]])

dim = 0, 可以这样考虑，将第一维压成1，将最外面的矩阵符号去掉，最后结果的维度是(2,4)：
[[3, 0, 0, 5], [2, 7, 6, 1]],
[[8, 2, 5, 6], [3, 4, 4, 8]],
[[7, 5, 5, 9], [2, 7, 3, 7]]

按照axis = 0去进行加法:
[[18,  7, 10, 20], [ 7, 18, 13, 16]]
"""

# dim = 1时变成横向压
"""
dim = 1, 可以这样考虑，将第二维压成1，去掉第二级括号，最后结果的维度是(3,4)：
"""
print("sum:", tensor1.sum(dim=1))

"""
dim = 2, 可以这样考虑，将第三维压成1，去掉第三级括号，最后结果的维度是(3,2)：
"""
print("sum:", tensor1.sum(dim=2))
```

```python
print("ave:", tensor1.mean(dim=1))
print("std:", tensor1.std())

print("max:", tensor1.max(dim=2))
print("min:", tensor1.min())

# argmin和argmax的序号就是完全扁平化后的顺序
print("argmin:", tensor1.argmin(dim=3))
print("argmax:", tensor1.argmax())
```
