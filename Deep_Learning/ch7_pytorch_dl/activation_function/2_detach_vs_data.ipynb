```python
import torch
```

```python
x1 = torch.tensor([1., 2., 3., 4.], requires_grad=True)
x2 = torch.tensor([1., 2., 3., 4.], requires_grad=True)
```

```python
y1 = x1.sigmoid()
y2 = x2.sigmoid()
```

```python
z1 = y1.data
z2 = y2.detach()

# 此时我们测试直接使用data和使用detach的区别
z1.zero_()
z2.zero_()
print(z1)
print(z2)
```

```python
y1.sum().backward()
y2.sum().backward()

print(x1.grad)
print(x2.grad)
```
