```python
import torch
import torch.nn as nn
```

```python
# linear layer
linear = nn.Linear(5,3)
```

```python
# 常数初始化
print(linear.weight)
nn.init.zeros_(linear.weight)
print(linear.weight)
```

```python
# linear.weight应该是5*3，但是pytorch默认使用转置来表示
nn.init.ones_(linear.weight)
print(linear.weight)
nn.init.constant_(linear.bias, 10)
print(linear.bias)
```

```python
# 秩初始化
nn.init.eye_(linear.weight)
print(linear.weight)
```

```python
# 均匀分布初始化
nn.init.uniform_(linear.weight, a=1, b=10)
print(linear.weight)
```

```python
# 正态分布初始化
nn.init.normal_(linear.weight, mean=0, std=1)
print(linear.weight)
```

```python
# Xavier
# 均匀分布
nn.init.xavier_uniform_(linear.weight)
print(linear.weight)

# 正态分布
nn.init.xavier_normal_(linear.weight)
print(linear.weight)
```

```python
# He
nn.init.kaiming_uniform_(linear.weight)
print(linear.weight)

nn.init.normal_(linear.weight)
print(linear.weight)
```
