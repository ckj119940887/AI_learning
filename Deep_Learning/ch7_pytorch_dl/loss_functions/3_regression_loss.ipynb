```python
import torch
import torch.nn as nn
```

```python
x = torch.randn(3, 5)
t = torch.randn(3, 5)
print(x)
print(t)
```

```python
# MAE(L1Loss)
loss = nn.L1Loss()
print(loss(x, t))
```

```python
# MSE(L2Loss)
loss = nn.MSELoss()
print(loss(x, t))
```

```python
# Smooth L1
loss = nn.SmoothL1Loss()
print(loss(x, t))
```
