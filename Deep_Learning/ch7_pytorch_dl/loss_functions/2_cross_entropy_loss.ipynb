```python
import torch
import torch.nn as nn
```

```python
# 1. 真实值为类别标签(0-7)
x = torch.randn(6, 8)
print(x)
```

```python
t = torch.tensor([1, 0, 3, 2, 5, 7])
print(t)
```

```python
loss = nn.CrossEntropyLoss()
print(loss(x, t))
```

```python
# 2. 真实值为分类概率或独热码
t = torch.randn(6, 8).softmax(dim=1)
print(t)
```

```python
print(loss(x, t))
```
