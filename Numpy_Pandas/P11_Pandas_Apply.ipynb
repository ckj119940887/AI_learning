```python
import numpy as np
import pandas as pd
from pandas import Series
```

```python
# apply 对DataFrame,Series进行逐行、逐列、逐数据的处理
s = pd.Series([1,2,3,4,5,6])
print(s * 10)
```

```python
def func(item):
    return item * 10

print(s.apply(func))
print(s.apply(lambda item: item * 10))

def func2(item, param):
    return item * param

# 这里在向func2传参时必须使用关键字传参，即指明参数名称
print(s.apply(func2, param=10))
```

```python
# 对DataFrame进行处理
df = pd.DataFrame({"a": [10, 20, 40], "b": [100, 200, 300]})
print(df)

# 接收的参数一定是Series
def func3(s):
    return s.apply(lambda item: item * 10)

def sum_col(s):
    return s.sum()

print(df.apply(func3, axis=1))
print(df.apply(sum_col, axis=0))
```

```python
# 向量化函数
# 将一个函数变成向量化函数
def f(x, y):
    return x + y

vec_f = np.vectorize(f)
print(vec_f(df.a, df.b))
```
