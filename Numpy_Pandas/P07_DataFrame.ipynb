```python
import pandas as pd
```

```python
mango_series = pd.Series([4,5,6,3,1])
apple_series = pd.Series([5,4,3,0,2])
banana_series = pd.Series([2,3,5,2,7])

# DataFrame是由多个Series组成的
# 可以用index指定行索引，用columns指定列的名字和顺序
df = pd.DataFrame({"mango": mango_series, "apple": apple_series, "banana": banana_series})
df
```

```python
# index
# columns
# value
# ndim
# shape
# size
# dtype
# name
# loc: 显示索引，切片
# iloc: 隐示索引，切片
# at: 访问单个元素
# iat
# T: 转置

print(df.index)
print(df.columns)
print(df.values)
print(df.ndim)
print(df.T)
print(df.at[1, "mango"])
```

```python
print(df.loc[0:2, "mango":"apple"])
print("---------------")
print(df.loc[[0,3], "banana"])
print("---------------")
print(df.loc[:, "mango":"apple"])
print("---------------")

print(df.iloc[:, 2])
```

```python
# head
# tail
# isin: 判断元素是否在Series
# isna: 判断是否为Nan
# np.sum
# np.mean
# np.max
# np.min
# np.std: 标准差
# np.var: 方差

print(df["banana"].sum())
```

```python
# bool索引
print(df[df["banana"] > 2])
```

```python
df["banana"] = df["banana"] * 2
print(df)
```

```python
# 修改操作
# 增加新的列
df["kejun"] = [1,2,3,4,5]
print(df)

# 取多列数据
# ["banana"] 本质上是一个列表，包含了多个列名，返回的是一个DataFrame对象
print(df[["banana"]])
print(type(df["banana"])) # Series
print(type(df[["banana"]])) # DataFrame
print("---------------")

# 修改index名
df.index = ["a", "b", "c", "d", "e"]
print(df)

# 修改列名
df.columns = ["A", "B", "C", "D"]
print(df)
```
