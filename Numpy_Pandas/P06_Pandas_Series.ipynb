```python
import pandas as pd
```

```python
# Series 很像一个一维数组，不过是标签化的一维数组，每个数据都有一个标签(索引)
# 此外还可以赋予name,进而使用name和标签来索引一个数据

s = pd.Series([1,2,3,4,5,6])
s
```

```python
s2 = pd.Series(data = [1,2,3,4,5,6], index = ['a','b','c','d','e','f'], name = 'test')
s2
```

```python
# 通过字典创建
dic = {"a":1, "b":2, "c":3, "d":4, "e":5}
s3 = pd.Series(dic)
s3
```

```python
# index
# value
# ndim
# shape
# size
# dtype
# name
# loc: 显示索引，切片
# iloc: 隐示索引，切片
# at
# iat

dic2 = {"a":1, "b":2, "c":3}
s4 = pd.Series(dic2)
print(s4.loc['a'])
print(s4.loc['a':'c'])
print(s4.iloc[1])
print(s4.iloc[0:2])
print(s4.at["a"])
print(s4.iat[1])
```

```python
# head
# tail
# isin: 判断元素是否在Series
# isna: 判断是否为Nan
# np.sum
# np.mean
# np.max
# np.min
# np.std: 标准差
# np.var: 方差
```

```python
# bool索引
dict3 = {"a":1, "b":2, "c":3}
s5 = pd.Series(dict3)
bools = s5 > s5.mean()
print(type(bools))
print(s5[bools])
```
