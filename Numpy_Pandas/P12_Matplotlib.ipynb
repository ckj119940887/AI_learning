```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建画布
plt.figure(figsize=(10,6))

# 创建2行1列的图，现在指向第一个图
plt.subplot(2,1,1)
# 设置x轴范围
plt.xlim(0, 10)
# 设置y轴范围
plt.ylim(-1, 1)
# 设置x轴label
plt.xlabel("x")
# 设置y轴label
plt.ylabel("sin(x)")
# 设置标题
plt.title("x-sin(x)")
plt.plot(x, y1)

# 创建2行1列的图，现在指向第一个图
plt.subplot(2,1,2)
# 设置x轴范围
plt.xlim(0, 10)
# 设置y轴范围
plt.ylim(-1, 1)
# 设置x轴label
plt.xlabel("x")
# 设置y轴label
plt.ylabel("cos(x)")
# 设置标题
plt.title("x-cos(x)")
plt.plot(x, y2)
```
