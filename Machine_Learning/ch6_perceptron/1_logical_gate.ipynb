```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    res = (x1*w1 + x2*w2)
    if res <= theta:
        return 0
    else:
        return 1
```

```python
import numpy as np
```

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    beta = np.array([0.5, 0.5])
    b = -0.7
    res = np.sum(beta * x) + b
    if res < 0:
        return 0
    else:
        return 1

print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    beta = np.array([-0.5, -0.5])
    b = 0.7
    res = np.sum(beta * x) + b
    if res < 0:
        return 0
    else:
        return 1

print(NAND(0, 0))
print(NAND(0, 1))
print(NAND(1, 0))
print(NAND(1, 1))
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    beta = np.array([0.5, 0.5])
    b = 0
    res = np.sum(beta * x) + b
    if res <= 0:
        return 0
    else:
        return 1

print(OR(0, 0))
print(OR(0, 1))
print(OR(1, 0))
print(OR(1, 1))
```

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print(XOR(0, 0))
print(XOR(0, 1))
print(XOR(1, 0))
print(XOR(1, 1))
```
