```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# confusion matrix
labels = ["猫", "狗"]
y_true = ["猫", "猫", "猫", "猫", "猫", "猫", "狗", "狗", "狗", "狗"] #真实值
y_pred1 = ["猫", "猫", "狗", "猫", "猫", "猫", "猫", "猫", "狗", "狗"] #预测值

matrix = confusion_matrix(y_true, y_pred1, labels=labels)

sns.heatmap(matrix, annot=True, fmt="g")
```

```python
# accuracy
accuracy = accuracy_score(y_true, y_pred1)
print(accuracy)
```

```python
# precision
precision = precision_score(y_true, y_pred1, pos_label="猫")
print(precision)
```

```python
# recall
recall = recall_score(y_true, y_pred1, pos_label="猫")
print(recall)
```

```python
# f1 score
# recall和precision的调和平均数
f1 = f1_score(y_true, y_pred1, pos_label="猫")
print(f1)
```
