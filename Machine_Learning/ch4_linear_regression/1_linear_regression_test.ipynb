```python
import matplotlib.pyplot as plt
from numpy.ma.core import reshape
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# 自变量，每周学习时长
# 每个数据都是一个list，表明每行对应多个特征
# 这本应该是经过train_test_split后的结果，只不过现在是我们手动给的
X = [[5], [8], [10], [12], [15], [3], [7], [9], [14], [6]]
# 因变量，数学考试成绩
y = [55, 65, 70, 75, 85, 50, 60, 72, 80, 58]

# 默认fit_intercept是True，也就是默认是有截距的
model = LinearRegression()
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
```

```python
x_test = [[11]]
model.predict(x_test)
```

```python
plt.scatter(X, y)

# -1 代表自己推断行的维度
x_line = np.arange(0, 18, 0.1).reshape(-1, 1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='red')
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
model = SGDRegressor(
    loss="squared_error", # 损失函数，默认为均方误差
    fit_intercept=True, # 是否计算偏置
    learning_rate="constant", # 学习率策略
    eta0=0.01, # 初始学习率
    max_iter=1000, # 最大迭代次数
    tol=1e-8, # (tol > best_loss - loss)时停止迭代
)

model.fit(X, y)
print(model.coef_)
print(model.intercept_)
```
