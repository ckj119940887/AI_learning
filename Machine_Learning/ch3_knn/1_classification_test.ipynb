```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
X = np.array([[2, 1], [3, 1], [1, 4], [2, 6]]) # 特征
y = np.array([0, 0, 1, 1]) # 标签

# model
knn = KNeighborsClassifier(n_neighbors=2)

# train
knn.fit(X, y)

# predict
print(knn.predict([[4, 9]]))

X1 = X[y == 0]
X2 = X[y == 1]

plt.scatter(X1[:, 0], X1[:, 1], c='red', marker='o', label='Class 0')
plt.scatter(X2[:, 0], X2[:, 1], c='blue', marker='o', label='Class 1')
plt.scatter(4, 9)
plt.show()
```
