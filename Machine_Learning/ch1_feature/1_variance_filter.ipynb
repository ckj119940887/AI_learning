```python
import numpy as np
```

```python
# np.random.randn, 标准正态分布：均值 0、标准差 1
# 形状用 size= 指定：normal(..., size=(2,3))
a = np.random.randn(100)
print(np.var(a))
```

```python
# 假设一个分布X~(u, sigma^2)
# 那么0.1X ~ (0.1u, 0.01 * sigma^2)
# b = np.random.randn(100) * 0.1
# print(np.var(b))

# 也可以直接使用normal来生成服从正态分布的数据
# 采样 一般正态分布：均值 loc，标准差 scale
# u = 5, sigma = 0.1
b = np.random.normal(loc=5, scale=0.1, size=100)
print(np.var(b))
```

```python
# 将两列数据堆叠在一起
# np.vstack 用来把多个数组沿着竖直方向拼接——也就是按“行”把它们叠起来（等价于在 axis=0 方向做 concatenate）
X = np.vstack([a,b]).T
print(X.shape)
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
# VarianceThreshold 是 scikit-learn 里一个最简单的特征选择方法：把方差太小（几乎不变化）的特征删掉，因为这类特征通常对模型没什么信息量。
# 将小于0.01的都过滤掉
# 注意上述代码中的数据是随机生成的，每次方差都是不一样的
# 创建过滤器
vt = VarianceThreshold(0.01)
# 过滤完后只剩下a,因为a的方差远大于0.01
X_new = vt.fit_transform(X)
print(X_new.shape)

print(X_new)
print(a)
```
