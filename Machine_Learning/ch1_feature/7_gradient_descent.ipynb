```python
# 目标函数
def f(x):
    return x ** 2

# 梯度函数
def gradient(x):
    return 2 * x
```

```python
# 初始化自变量和学习率
x = 1
alpha = 0.1

x_list = []
y_list = []

for i in range(100):
    x_list.append(x)
    y_list.append(f(x))

    # 计算梯度
    grad = gradient(x)
    # 更新参数
    x = x - alpha * grad

print(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1, 1, 0.01)

plt.plot(x, f(x))
plt.plot(x_list, y_list, 'r')
plt.scatter(x_list, y_list, color='red')
plt.show()
```
