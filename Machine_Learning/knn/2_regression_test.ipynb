```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
```

```python
# data
X = np.array([[2, 1], [3, 1], [1, 4], [2, 6]]) # 特征
y = np.array([0.5, 1.5, 4, 3.2]) # 标签

# model
knn = KNeighborsRegressor(n_neighbors=2)

# train
knn.fit(X, y)

print(knn.predict([[4, 9]]))
```
