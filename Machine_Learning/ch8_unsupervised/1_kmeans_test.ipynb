```python
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs # 自动生成聚集分布的点
from sklearn.metrics import silhouette_score, calinski_harabasz_score
```

```python
# y代表的是标签，由于我们是无监督学习，所以y没有用
X, y = make_blobs(n_samples=300, centers=3, cluster_std=2, random_state=42)

fig, ax = plt.subplots(2, figsize=(10,10))
ax[0].scatter(X[:, 0], X[:, 1], c='gray')
ax[0].set_title("original dataset")

model = KMeans(n_clusters=3)
model.fit(X)

# 获取簇的中心点
centers = model.cluster_centers_

y_pred = model.predict(X)

ax[1].scatter(X[:, 0], X[:, 1], c=y_pred)
ax[1].scatter(centers[:, 0], centers[:, 1], c='red', marker='o')
ax[1].set_title("kmeans")
```

```python
# metrics
print(model.inertia_) # 簇内平方和
print(silhouette_score(X, y_pred)) # 轮廓系数
print(calinski_harabasz_score(X, y_pred)) # CH指数
```
